### Teste da função quadrática

```python
# Teste de função quadrática

x = np.linspace(-10, 10, int((10 + 10)*1000))
y = float(1)*x**2 + float(1)*x + float(1)

# Processamento de dados
fig, ax = plt.subplots()
ax.plot(x, y)

# Saída de dados
plt.show()
```

```python
# Definindo função quadrática

def quadratica(intervalo: list, coef: list) -> list:
    """ Esta função gera uma lista de atributos para plotagem de uma função quadrática """
    x = np.linspace(intervalo[0], intervalo[1], int((intervalo[1]-intervalo[0])*1000))
    y = float(coef[0])*x**2 + float(coef[1])*x + float(coef[2])
    return [x,  y]
```

```python
# teste para plotagem

q = quadratica([-10, 10], [1,1,1])

# processando atributos para plotagem
fig, ax = plt.subplots()
ax.plot(q[0], q[1])

# Saída de dados
plt.show()
```

### Teste da função afim

```python
# Teste de função afim

x = np.linspace(-10, 10, int((10 + 10)*1000))
y = float(1)*x + float(1)

# Processamento de dados
fig, ax = plt.subplots()
ax.plot(x, y)

# Saída de dados
plt.show()
```

```python
# Definindo função afim

def afim(intervalo: list, coef: list) -> list:
    """ Esta função gera uma lista de atributos para plotagem de uma função afim """
    x = np.linspace(intervalo[0], intervalo[1], int((intervalo[1]-intervalo[0])*1000))
    y = float(coef[0])*x + float(coef[1])
    return [x, y]
```

```python
# teste para plotagem

a = afim([1, 10], [1,1])

# processando atributos para plotagem
fig, ax = plt.subplots()
ax.plot(a[0], a[1])

# Saída de dados
plt.show()
```

### Teste da função logarítmica

```python
# Teste de função log

x = np.linspace(1, 10, int((10 - 1)*1000))
y = np.log(x)

# Processamento de dados
fig, ax = plt.subplots()
ax.plot(x, y)

# Saída de dados
plt.show()
```

```python
# Definindo função logarítmica

def logar(intervalo: list) -> list:
    """ Esta função gera uma lista de atributos para plotagem de uma função logarítmica """
    x = np.linspace(intervalo[0], intervalo[1], int((intervalo[1]-intervalo[0])*1000))
    y = np.log(x)
    return [x, y]
```

```python
# teste para plotagem

l = logar([1, 10])

# processando atributos para plotagem
fig, ax = plt.subplots()
ax.plot(l[0], l[1])

# Saída de dados
plt.show()
```

### Teste da função exponencial

```python
# Teste de função exp

x = np.linspace(1, 10, int((10 - 1)*1000))
y = np.exp(x)

# Processamento de dados
fig, ax = plt.subplots()
ax.plot(x, y)

# Saída de dados
plt.show()
```

```python
# Definindo função exponencial

def exp(intervalo: list) -> list:
    """ Esta função gera uma lista de atributos para plotagem de uma função exponencial """
    x = np.linspace(intervalo[0], intervalo[1], int((intervalo[1]-intervalo[0])*1000))
    y = np.exp(x)
    return [x, y]
```

```python
# teste para plotagem

e = exp([1, 10])

# processando atributos para plotagem
fig, ax = plt.subplots()
ax.plot(e[0], e[1])

# Saída de dados
plt.show()
```

## 1. Requisito 1b

Plotar o gráfico

função impressora(grafico: objeto gráfico) -> null

Esta função imprime na tela o gráfico da função escolhida pelo usuário.

```python
# Importando bibliotecas

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Teste de funções de gráfico

dados = [[1,2,3,4], [1,4,2,3]]

fig, ax = plt.subplots()
grafico = ax.plot(dados[0], dados[1])
plt.show()

print(type(grafico))
```

```python
# Teste de plot

import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.subplots()
ax.plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])
```

```python
# definindo a função impressora do gráfico

def impressora(x, y) -> None:
    """Função para plotagem da função"""
    fig, grafico = plt.subplots()
    grafico.plot(x,y)
    plt.show()
```

```python
# Teste de função impressora

x = np.linspace(10, 20, int((20-10)*1000))
y = np.exp(x)

impressora(x, y)
```

## 2. Requisito 2

Menu com as quatro funções a serem plotadas

função leitor_operacao() -> string
Imprime um menu na tela contendo as funções afim, quadrática, logarítmica e exponencial para o usuário escolher qual delas plotar.


```python
# Definindo a função leitora da operação 


def leitora_operacao() -> str:
    """Esta funcao le a função a ser realizada."""
    operacao = input("""Digite o tipo de função que deseja plotar. 
Digite quad para quadratica (a*x2 + b*x + c); 
afim para afim (a*x + b); 
log para logarítmica (log(x)); 
exp para exponencial (exp(x)): \n""")
    return operacao
```

```python
# Teste de função para armazenamento do dado

x = leitora_operacao()
```

```python
# Teste de armazenamento do dado

print(x)
```

## 3. Requisito 3

Inserir coeficientes da função afim e quadrática

função leitor_coeficiente(nome_funcao: string) -> lista 
Lê os coeficientes no caso de o usuário escolher plotar uma função afim ou quadrática.


```python
# Definindo a função leitora de coeficientes 

def leitora_coeficientes(operacao: str) -> list:
    """Função para inserir o coeficientes da função"""
    a = input("Qual o coeficiente a? ")
    b = input('Qual o coeficiente b? ')
    if operacao == 'quad':
        c = input('Qual o coeficiente c? ')
        return [a, b, c]
    else:
        return [a, b]
```

```python
# Teste da função com a quadrática

q = leitora_coeficientes('quad')

print(q)
```

```python
# Teste da função com a afim

a = leitora_coeficientes('afim')

print(a)
```

## 4. Intervalo de X

Inserir o intervalo de X no gráfico

função leitora_int() -> list
Lê o intervalo de x a ser apresentado no gráfico.

```python
# Definindo a função leitora de intervalo x 

def leitora_int() -> list:
    """Função para inserir o intervalo de x da função"""
    inicio = float(input('Qual o valor de x inicio da funcao? '))
    fim = float(input('Qual o valor de x final da função? '))
    return [inicio, fim]
                
```

```python
# Teste da função

i = leitora_int()
```

```python
# Teste de armazenamento de dado com a função

print(i)
```

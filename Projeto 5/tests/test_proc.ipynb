### Definindo a fun√ß√£o  get_web

```python
# Definindo a fun√ß√£o get_web() -> arquivo

def get_web(dados: list):
    """ esta fun√ß√£o l√™ e baixa os dados na p√°gina web requerida """
    arquivo = 'pagina.txt'
    if dados[1] == 'json':
        r = requests.get(dados[0])
        page = r.json()
        with open(arquivo, 'w') as f:
            f.write(json.dumps(page))
    elif dados[1] == 'raw':
        r = requests.get(dados[0])
        page = r.content
        with open(arquivo, 'wb') as fd:
            fd.write(page)
    else:
        r = requests.get(dados[0])
        page = r.text
        with open(arquivo, 'w', encoding='utf-8') as f:
            f.write(page)
    return arquivo
```

```python
# Testando a fun√ß√£o get_web

## Para json
get_web(['https://api.github.com/events', 'json'])

## Para raw
get_web(['https://api.github.com/events', 'raw'])

## Para c√≥digo fonte
get_web(['https://api.github.com/events', 'codigo'])
```

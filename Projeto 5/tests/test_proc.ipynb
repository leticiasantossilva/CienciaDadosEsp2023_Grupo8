# Módulo de processamento

Este módulo prevê a função de processamento dos dados no Aplicativo de Coleta de Dados na Web. A função utiliza a biblioteca Python Requests e deverá ser capaz de ler e baixar dados de páginas web.

## Requisito 2

Baixar a página da web requerida pelo usuário.

Função get_web(nome_pagina) -> file 

Atendendo ao requito 2 do aplicativo, esta função que deve ler e baixar os dados na página web requerida.

### Transformando dados baixados em arquivos

```python
# Salvando em arquivo dados coletados em json

with open('j_test.txt', 'w') as f:
            f.write(json.dumps(page_j))
        
```

### Definindo a função  get_web

```python
# Definindo a função get_web() -> arquivo

def get_web(dados: list):
    """ esta função lê e baixa os dados na página web requerida """
    arquivo = 'pagina.txt'
    if dados[1] == 'json':
        r = requests.get(dados[0])
        page = r.json()
        with open(arquivo, 'w') as f:
            f.write(json.dumps(page))
    elif dados[1] == 'raw':
        r = requests.get(dados[0])
        page = r.content
        with open(arquivo, 'wb') as fd:
            fd.write(page)
    else:
        r = requests.get(dados[0])
        page = r.text
        with open(arquivo, 'w', encoding='utf-8') as f:
            f.write(page)
    return arquivo
```

```python
# Testando a função get_web

## Para json
get_web(['https://api.github.com/events', 'json'])

## Para raw
get_web(['https://api.github.com/events', 'raw'])

## Para código fonte
get_web(['https://api.github.com/events', 'codigo'])
```

### Exemplo da biblioteca Scipy

Referência: [Função interp1d](https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.interp1d.html#scipy.interpolate.interp1d)

```python
# Teste de interpolate.interp1d(x, y)
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

# Atribuição de variáveis para a função de interpolação

x = np.arange(0, 10)  # criando objeto de tipo numpy.ndarray
y = np.exp(-x/3.0)    # função exponencial do Numpy - mais rápida do que a nativa do Python

# Define função de intepolação das variáveis
f = interpolate.interp1d(x, y)
```

```python
# Aplicando a função f() de interpolação

xnew = np.arange(0, 9, 0.1)
ynew = f(xnew)    # use o retorno da função de interpolação

# Apresenta o gráfico da função
plt.plot(x, y, 'o', xnew, ynew, '-')
plt.show()
```

## Define a função ajusta_curva()

```python
# Importação

from collections.abc import Callable
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt

# Definindo a função
def ajusta_curva(pontos: list) -> Callable:
    """ Esta função retorna as funções de interpolação """
    return interpolate.interp1d(pontos[0], pontos[1])   

# Pontos[0] é a lista de todos os X, ordenados; e Pontos[1] é lista que 
```

```python
#testes

x = np.arange(5, 30, 5)
y = x

f = interpolate.interp1d(x, y)

xn = np.arange(5, 25, 0.1)
yn = f(xn)

plt.plot(x, y, 'ro', xn, yn, 'g-')
plt.show()
```

```python
# Teste de interpolate.interp1d(x, y)
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

# Atribuição de variáveis para a função de interpolação

x = np.arange(0, 10)  # criando objeto de tipo numpy.ndarray
y = np.exp(-x/3.0)    # função exponencial do Numpy - mais rápida do que a nativa do Python

# Define função de intepolação das variáveis
f = interpolate.interp1d(x, y)
```

```python
# Aplicando a função f() de interpolação

xnew = np.arange(0, 9, 0.1)
ynew = f(xnew)    # use o retorno da função de interpolação
```

```python
# Apresentando o gráfico da função

plt.plot(x, y, 'o', xnew, ynew, 'b-')
plt.show()
```

### Função de interpolação

```python
# Definindo função 

def interp(dados: list) -> list:
    """ Esta função aplica interpolação na lista de pontos inserida pelo usuário, gerando uma lista com elementos np.ndarray """
    x1 = [float(i) for i in dados[0]]
    y1 = [float(i) for i in dados[1]]
    
    x = np.array(x1)
    y = np.array(y1)
    
    f = interpolate.interp1d(x, y)
    
    xnew = np.arange(min(x), max(x), 0.0001)
    ynew = f(xnew)   # use interpolation function returned by `interp1d`
    
    return (x, y, xnew, ynew)
```

```python
# Teste da função

test = interp([[1, 5], [3, 4]])

print(test)
```

```python

# Teste de saída em gráfico

plt.plot(test[0], test[1], 'o', test[2], test[3], 'b-')
plt.show()
```
